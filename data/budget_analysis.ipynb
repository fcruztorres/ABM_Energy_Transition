```python
import pandas as pd
import numpy as np
```

```python
budget_data = pd.read_csv("budget_data.csv", sep=";", skiprows=6, index_col=0)
municipalities = pd.read_csv("municipalities.csv")
```

```python
budget_data = budget_data.transpose()
budget_data = budget_data.drop(["Task field / balance sheet item"])
budget_data = budget_data.apply(pd.to_numeric, errors='coerce')
```

```python
budget_data.columns = ['Alblasserdam', 'Albrandswaard', 'Alphen aan den Rijn', 'Barendrecht',
       'Bodegraven-Reeuwijk', 'Brielle', 'Capelle aan den IJssel', 'Delft',
       'Dordrecht', 'Goeree-Overflakkee', 'Gorinchem', 'Gouda',
       "'s-Gravenhage", 'Hardinxveld-Giessendam',
       'Hellevoetsluis', 'Hendrik-Ido-Ambacht', 'Hillegom', 'Hoeksche Waard',
       'Kaag and Braassem', 'Katwijk', 'Krimpen aan den IJssel',
       'Krimpenerwaard', 'Lansingerland', 'Lead', 'Leiderdorp',
       'Leidschendam-Voorburg', 'Lisse', 'Maassluis', 'Midden-Delfland',
       'Molenlanden', 'Nieuwkoop', 'Nissewaard', 'Noordwijk', 'Oegstgeest',
       'Papendrecht', 'Pijnacker-Nootdorp', 'Ridderkerk', 'Rijswijk',
       'Rotterdam', 'Schiedam', 'Sliedrecht', 'Teylingen', 'Vlaardingen',
       'Advances', 'Waddinxveen', 'Wassenaar', 'Westland', 'Westvoorne',
       'Zoetermeer', 'Zoeterwoude', 'Zuidplas', 'Zwijndrecht']
```

```python
mrdh_budget = budget_data[municipalities.name.values]
```

```python
mrdh_budget = mrdh_budget.transpose()
mrdh_budget = mrdh_budget * 1000
```

```python
mrdh_budget["Energy spendings"] = mrdh_budget["0.3 Management of other buildings and land"] + mrdh_budget["3.1 Economic development"] + mrdh_budget["3.2 Physical business infrastructure"] +  mrdh_budget["7.4 Environmental management"] + mrdh_budget["8.2 Land development (non-commercial .."]
```

```python
mrdh_budget
```

```python
municipalities = municipalities.merge(mrdh_budget["Energy spendings"], how = "left", left_on="name", right_index=True)
```

```python
municipalities["allocated funds"] = municipalities["Energy spendings"]
municipalities
```

```python
municipalities.iloc[0:,0:6].to_csv("municipalities.csv", index = False)
```
